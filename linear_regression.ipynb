```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error,r2_score
diabetes_x,diabetes_y=datasets.load_diabetes(return_X_y=True)
diabetes_x=diabetes_x[:,np.newaxis,2]
diabetes_x_train=diabetes_x[:-20]
diabetes_x_test=diabetes_x[-20:]
diabetes_y_train=diabetes_y[:-20]#split the depedent variable
diabetes_y_test=diabetes_y[-20:]
reg=linear_model.LinearRegression()#craete an object of the model
reg.fit(diabetes_x_train,diabetes_y_train)
y_pred=reg.predict(diabetes_x_test)
print(y_pred)
plt.scatter(diabetes_x_test,diabetes_y_test,c="red",marker="*",)
plt.plot(diabetes_x_test,y_pred,color= "blue",linestyle="dotted",marker="<",mfc="green",mec="yellow")
plt.grid()
plt.title("Diabetes")
plt.show()
```

```python
print(diabetes_x.shape)
```

```python
print("r2 score:",r2_score(diabetes_y_test,y_pred))
```
